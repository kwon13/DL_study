```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# 2개의 독립변수가 있는 데이터 셋
data=[[2,0,81],[4,4,93],[6,2,91],[8,3,97]]
x1=[i[0] for i in data]
x2=[i[1] for i in data]
y=[i[2] for i in data]
```

```python
# 그래프로 데이터 확인
ax=plt.axes(projection='3d')
ax.scatter(x1,x2,y)
plt.show()
```

```python
# 넘파이 배열로 변환
x1_data=np.array(x1)
x2_data=np.array(x2)
y_data=np.array(y)

# 가중치와 절편 초기화 
w1=0
w2=0
b=0

# 학습률과 학습 반복횟수 설정
lr=0.02
epochs=2001

for i in range(epochs):
    # 예측값과 오차값 계산
    y_hat=w1 * x1_data + w2 * x2_data + b
    error=y_data - y_hat
    
    # 미분 값 계산
    w1_diff=-(2/len(x1_data)) * sum(x1_data * error)
    w2_diff=-(2/len(x2_data)) * sum(x2_data * error)

    b_diff=-(2/len(x1_data)) * sum(error)

    # 가중치와 절편 업데이트
    w1=w1-lr*w1_diff
    w2=w2-lr*w2_diff
    b=b-lr*b_diff

    if i % 100 == 0:
        print(' epoch: {}, 기울기: w1={:.2f} / w2={:.2f}, 절편: {:.2f}'.format(i,w1,w2,b))
```

```python
y_pred=w1 * x1_data + w2 * x2_data + b
plt.scatter(x1,y)
plt.plot(x1, y_pred, 'ro-')
plt.ylim(30,100)
plt.show()
```
